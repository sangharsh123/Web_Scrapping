```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

##  Find the URL 

```python
links='https://www.walmart.com/reviews/product/14898365?sort=submission-desc'
```

## Get the source HTML code as text

```python
page=requests.get(links)
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

## Convert that text into a bs4 lxml object 

###  Decide/Identify the tags

### Get the value(s) present in each tag into a list/tuple/dictionary

```python
name= soup.find_all('span', class_="review-footer-userNickname")
name
```

```python
len(name)
```

```python
cust_name=[]
for i in range(0, len(name)):
    cust_name.append(name[i].get_text())
cust_name=pd.Series(cust_name,name='cust_name')
```

```python
cust_name
```

```python
title=soup.findAll(class_="review-title font-bold")
title
```

```python
len(title)
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title=pd.Series(review_title,name='review_title')    
```

```python
review_title
```

```python
rating=soup.find_all("span", {"class":"stars-container"})
len(rating)
```

```python
review_stars = soup.find_all("div",class_="MeterInline")
```

```python
star_rate=[]
for i in range(0, len(review_stars)):
    star_rate.append(review_stars[i].get_text())
star_rate=pd.Series(star_rate, name='star_rate')
```

```python
star_rate
```

```python
review=soup.select("p", class_="review-text")
```

```python
review_body=[]
for i in range(0, len(review)):
    review_body.append(review[i].get_text())
   
```

```python
review_description=pd.Series(review_body,name='review_description')
```

```python
review_date = soup.find_all ('span',{"class":"review-date-submissionTime"})
len(review_date)
```

```python
reviews_date =[]
for i in range(0, len(review_date)):
    reviews_date.append(review_date[i].get_text())
 
```

```python
reviews_date=pd.Series(reviews_date, name='reviews_date')
reviews_date
```

## Create a data frame out of the final list

```python
data=pd.concat([cust_name,review_title,review_description,star_rate,reviews_date], axis=1)
data
```

## save csv file in excel format

```python
data.to_csv(r'E:\Review.csv', index=True)
```
